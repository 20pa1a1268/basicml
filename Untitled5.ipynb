```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
df=pd.read_csv("Iris.csv")
df.head()
```

```python
features=['SepalLengthCm',	'SepalWidthCm',	'PetalLengthCm',	'PetalWidthCm']
x=df.loc[:,features].values
y=df.loc[:,['Species']].values
```

```python
x=StandardScaler().fit_transform(x)
x[0:5]
```

```python
pca=PCA(n_components=2)
principalComponents=pca.fit_transform(x)
df_pca=pd.DataFrame(data=principalComponents,columns=['principal component 1','principal component 2'])
df_pca.head()
```

```python
finalDf=pd.concat([df_pca,df['Species']],axis=1)
finalDf.head()
```

```python
from sklearn.naive_bayes import GaussianNB
model=GaussianNB()
model=model.fit(df_pca,y.ravel())
y_predict=model.predict(df_pca)
Actual_predicted=pd.DataFrame({'predict':y_predict.flatten(),'Actual':y.flatten()})
Actual_predicted.head()
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y,y_predict)
```
